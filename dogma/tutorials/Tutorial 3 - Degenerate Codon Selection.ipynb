```python
from dogma import GeneticCode
from dogma.extensions import CodonSelector
```

```python
supE = GeneticCode(ncbi_id=1, updated_mappings={'TAG': 'Q'})
```

```python
must_exclude = ['*', 'C']
```

```python
cs = CodonSelector(
    genetic_code=supE,
    must_exclude=must_exclude
)
```

```python
cs.excluded_amino_acids
```

```python
len(cs.filtered_table)
```

```python
cs.filtered_table.sort_values('Num_uAAs', ascending=False).head(10)
```

```python
eq = cs.filtered_table[cs.filtered_table.EquimolarAA]
eq.sort_values('Num_uAAs', ascending=False).head(10)
```
